```python
#importing Libraries
import nltk
from nltk import word_tokenize
#Input data
text=""" There is a man on the hill, and I watched him with my telescope. There is a man on the hill, and 
he has a telescope. Iâ€™m on a hill, and I saw a man using my telescope."""
#word Tokenize
words=word_tokenize(text)
print(len(words))
print(words)
['There', 'is', 'a', 'man', 'on', 'the', 'hill', ',', 'and', 'I', 'watched', 'him', 'with', 'my', 'telescope', '.', 'There', 'is', 'a', 
'man', 'on', 'the', 'hill', ',', 'and', 'he', 'has', 'a', 'telescope', '.', 'I', 'â€™', 'm', 'on', 'a', 'hill', ',', 'and', 'I', 'saw', 'a', 
'man', 'using', 'my', 'telescope', '.']
# Frequency Distribution
from nltk.probability import FreqDist
fdist=FreqDist(words)
fdist.most_common(10)
# Plot graph
import matplotlib.pyplot as plt
fdist.plot(20)
```

```python
import re
def test(string):
    merged = re.split(r"([ ,!]+)", string)
    return [merged[::2], merged[1::2]]
s = input("Enter a string with , separators")
print("\nOriginal string:",s)
print("Split the said string into 2 lists: words and separators:")
print(test(s))
```
