```python
INS_COST, DEL_COST, SUB_COST = 1, 1, 1

def find_minimum_edit_distance(source, target):
    m, n = len(source), len(target)
    dp = [[0] * (m + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        dp[i][0] = dp[i - 1][0] + INS_COST

    for j in range(1, m + 1):
        dp[0][j] = dp[0][j - 1] + DEL_COST

    operations = []
    i, j = n, m

    while i > 0 and j > 0:
        if source[j - 1] == target[i - 1]:
            i, j = i - 1, j - 1
        else:
            if dp[i][j] == dp[i - 1][j - 1] + SUB_COST:
                operations.append(('SUBSTITUTE', source[j - 1], target[i - 1]))
                i, j = i - 1, j - 1
            elif dp[i][j] == dp[i - 1][j] + INS_COST:
                operations.append(('INSERT', target[i - 1]))
                i -= 1
            else:
                operations.append(('DELETE', source[j - 1]))
                j -= 1

    while j > 0:
        operations.append(('DELETE', source[j - 1]))
        j -= 1

    while i > 0:
        operations.append(('INSERT', target[i - 1]))
        i -= 1

    operations.reverse()
    return dp[n][m], operations

if __name__ == "__main__":
    source = input("Enter the source string: ").strip()
    target = input("Enter the target string: ").strip()

    distance, operations = find_minimum_edit_distance(source, target)
    
    insertions = sum(1 for op in operations if op[0] == 'INSERT')
    deletions = sum(1 for op in operations if op[0] == 'DELETE')
    substitutions = sum(1 for op in operations if op[0] == 'SUBSTITUTE')

    print(f"Minimum edit distance: {distance}")
    print(f"Number of insertions: {insertions}")
    print(f"Number of deletions: {deletions}")
    print(f"Number of substitutions: {substitutions}")
    print(f"Total number of operations: {insertions + deletions + substitutions}")
    print("Actual Operations:")
    
    for i, op in enumerate(operations, start=1):
        if op[0] == 'INSERT' or op[0] == 'DELETE':
            print(f"{i}) {op[0]}: {op[1]}")
        else:
            print(f"{i}) {op[0]}: {op[1]} by {op[2]}")
```
